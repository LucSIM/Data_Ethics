```python
import pandas as pd
import numpy as np
import seaborn as sns
from scipy.stats import skew
from collections import Counter

import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('Strategic_Subject_List.csv')
```

```python
plt.figure(figsize=(30,30))
sns.heatmap(data.corr(), square=True, annot=True,robust=True, yticklabels=1)
```

```python
missing_values = data.isnull().sum().sort_values(ascending = False)
percentage_missing_values = (missing_values/len(data))*100
pd.concat([missing_values, percentage_missing_values], axis = 1, keys= ['Missing values', '% Missing'])
```

```python
x = list(set(data['DRUG I']))
y = list(set(data['RACE CODE CD']))
z = list(set(data['AGE GROUP']))
print(z);

data['AGE GROUP'] = data['AGE GROUP'].map({'less than 20': 7, '20-30': 6, '30-40': 5, '40-50': 4, '50-60': 3, '60-70': 2, '70-80': 1})
data['SEX CODE CD'] = data['SEX CODE CD'].map({'F': 0, 'M': 1, 'X': 2})
data['RACE CODE CD'] = data['RACE CODE CD'].map({'WBH':0, 'I':1, 'WWH':2, 'U':3, 'BLK':4, 'API':5, 'WHI':6})
data['WEAPON I'] = data['WEAPON I'].map({'N': 0, 'Y': 1})
data['DRUG I'] = data['DRUG I'].map({'N': 0, 'Y': 1})
data['CPD ARREST I'] = data['CPD ARREST I'].map({'N': 0, 'Y': 1})


var = ['SSL SCORE', 'PREDICTOR RAT ARRESTS VIOLENT OFFENSES','PREDICTOR RAT NARCOTIC ARRESTS','PREDICTOR RAT VICTIM SHOOTING INCIDENTS',
       'PREDICTOR RAT UUW ARRESTS','PREDICTOR RAT VICTIM BATTERY OR ASSAULT','PREDICTOR RAT TREND IN CRIMINAL ACTIVITY',
       'PREDICTOR RAT GANG AFFILIATION','SEX CODE CD','RACE CODE CD','AGE GROUP','WEAPON I','DRUG I','CPD ARREST I']
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(data[var].corr(), square=True, annot=True,robust=True, yticklabels=1)
```
